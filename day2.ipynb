```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,np.pi,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
plt.plot(x,y)
plt.title("Graph of Sin Function")
plt.xlable("x-axis")
plt.ylable("y-axis")
plt.show()
```
