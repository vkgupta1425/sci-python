```python
import numpy as np 

np.sin(np.pi)
```

```python
import matplotlib.pyplot as plt 

x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x))
plt.title("The graph of sine")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$\sin(\theta)$")
plt.show()
```

```python
x = np.linspace(1, 10, 100)

y = (np.sin(x))**2 + np.cos(x)
y1 = np.sin(x)

plt.plot(x, y, color='r')
plt.plot(x, y1, color='cyan')
plt.show()
```
